En este cuaderno de Jupyter se realizó la carga de las 4 tablas correspondientes a las Streamings:
Amazon, Disney, Hulu y Netflix, así como el proceso de ETL.

```python
# Importación de la librería a usar en el proceso

import pandas as pd
```

```python
# Carga de la tabla de Amazoon Prime, así como una mirada a los primeros dos registros para ir analizando
# los datos

df_amazon = pd.read_csv ('./Datasets/amazon_prime_titles-score.csv')
df_amazon.head(2)
```

```python
df_amazon.info()
```

```python
# Carga de la tabla de Disney Plus

df_disney = pd.read_csv ('./Datasets/disney_plus_titles-score.csv')
df_disney.head(2)
```

```python
df_disney.info()
```

```python
# Carga de la tabla de Hulu

df_hulu = pd.read_csv ('./Datasets/hulu_titles-score (2).csv')
df_hulu.head(2)
```

```python
df_hulu.info()
```

```python
# Carga de la tabla de Netflix

df_netflix = pd.read_csv ('./Datasets/netflix_titles-score.csv')
df_netflix.head(2)
```

```python
df_netflix.info()
```

```python
# Generar campo id en todas las tablas

df_amazon ['id'] = 'a' + df_amazon ['show_id']
df_disney ['id'] = 'd' + df_disney ['show_id']
df_hulu ['id'] = 'h' + df_hulu ['show_id']
df_netflix ['id'] = 'n' + df_netflix ['show_id']
```

```python
# Reemplazar los valores nulos en la columna 'rating' por 'G'

df_amazon['rating'].fillna('G', inplace = True)
df_disney['rating'].fillna('G', inplace = True)
df_hulu['rating'].fillna('G', inplace = True)
df_netflix['rating'].fillna('G', inplace = True)
```

```python
# Convertir los registros de la columna 'date_added' a formato AAAA-mm-dd

df_amazon['date_added'] = pd.to_datetime(df_amazon['date_added'])
df_disney['date_added'] = pd.to_datetime(df_disney['date_added'])
df_hulu['date_added'] = pd.to_datetime(df_hulu['date_added'])
df_netflix['date_added'] = pd.to_datetime(df_netflix['date_added'])
```

```python
# Convertir la columna 'cast' del Dataframe 'df_hulu' en object (ya que se cargó como float)

df_hulu['cast'] = df_hulu['cast'].astype(object)
```

```python
# Convertir a minúsculas todas las columnas de texto en cada tabla

df_amazon['type'] = df_amazon['type'].str.lower()
df_amazon['title'] = df_amazon['title'].str.lower()
df_amazon['director'] = df_amazon['director'].str.lower()
df_amazon['cast'] = df_amazon['cast'].str.lower()
df_amazon['country'] = df_amazon['country'].str.lower()
df_amazon['rating'] = df_amazon['rating'].str.lower()
df_amazon['duration'] = df_amazon['duration'].str.lower()
df_amazon['listed_in'] = df_amazon['listed_in'].str.lower()
df_amazon['description'] = df_amazon['description'].str.lower()

df_disney['type'] = df_disney['type'].str.lower()
df_disney['title'] = df_disney['title'].str.lower()
df_disney['director'] = df_disney['director'].str.lower()
df_disney['cast'] = df_disney['cast'].str.lower()
df_disney['country'] = df_disney['country'].str.lower()
df_disney['rating'] = df_disney['rating'].str.lower()
df_disney['duration'] = df_disney['duration'].str.lower()
df_disney['listed_in'] = df_disney['listed_in'].str.lower()
df_disney['description'] = df_disney['description'].str.lower()

df_hulu['type'] = df_hulu['type'].str.lower()
df_hulu['title'] = df_hulu['title'].str.lower()
df_hulu['director'] = df_hulu['director'].str.lower()
df_hulu['cast'] = df_hulu['cast'].str.lower()
df_hulu['country'] = df_hulu['country'].str.lower()
df_hulu['rating'] = df_hulu['rating'].str.lower()
df_hulu['duration'] = df_hulu['duration'].str.lower()
df_hulu['listed_in'] = df_hulu['listed_in'].str.lower()
df_hulu['description'] = df_hulu['description'].str.lower()

df_netflix['type'] = df_netflix['type'].str.lower()
df_netflix['title'] = df_netflix['title'].str.lower()
df_netflix['director'] = df_netflix['director'].str.lower()
df_netflix['cast'] = df_netflix['cast'].str.lower()
df_netflix['country'] = df_netflix['country'].str.lower()
df_netflix['rating'] = df_netflix['rating'].str.lower()
df_netflix['duration'] = df_netflix['duration'].str.lower()
df_netflix['listed_in'] = df_netflix['listed_in'].str.lower()
df_netflix['description'] = df_netflix['description'].str.lower()
```

```python
# Dividiendo la columna 'duration' en 'duration_int' y 'duration_type' en la tabla de Amazon

duration = df_amazon['duration'].str.split(expand = True)
duration.columns = ['duration_int', 'duration_type']
df_amazon = pd.concat ([df_amazon, duration], axis = 1)


# Cambiando el tipo de dato en 'duration_int' a int
df_amazon['duration_int'] = df_amazon['duration_int'].astype(int)

# Normalizando los registros 'seasons' en 'season'
df_amazon['duration_type'].replace('seasons', 'season', inplace = True)
```

```python
# Dividiendo la columna 'duration' en 'duration_int' y 'duration_type' en la tabla de Disney

duration = df_disney['duration'].str.split(expand = True)
duration.columns = ['duration_int', 'duration_type']
df_disney = pd.concat ([df_disney, duration], axis = 1)


# Cambiando el tipo de dato en 'duration_int' a int
df_disney['duration_int'] = df_disney['duration_int'].astype(int)

# Normalizando los registros 'seasons' en 'season'
df_disney['duration_type'].replace('seasons', 'season', inplace = True)
```

```python
# Dividiendo la columna 'duration' en 'duration_int' y 'duration_type' en la tabla de Hulu

duration = df_hulu['duration'].str.split(expand = True)
duration.columns = ['duration_int', 'duration_type']
df_hulu = pd.concat ([df_hulu, duration], axis = 1)

# Cambiando el tipo de dato en 'duration_int' a int
df_hulu['duration_int'] = df_hulu['duration_int'].astype(float).astype('Int32')

# Normalizando los registros 'seasons' en 'season'
df_hulu['duration_type'].replace('seasons', 'season', inplace = True)
```

```python
# Dividiendo la columna 'duration' en 'duration_int' y 'duration_type' en la tabla de Netflix

duration = df_netflix['duration'].str.split(expand = True)
duration.columns = ['duration_int', 'duration_type']
df_netflix = pd.concat ([df_netflix, duration], axis = 1)

# Cambiando el tipo de dato en 'duration_int' a int
df_netflix['duration_int'] = df_netflix['duration_int'].astype(float).astype('Int32')

# Normalizando los registros 'seasons' en 'season'
df_netflix['duration_type'].replace('seasons', 'season', inplace = True)
```

```python
# Se unen todas las tablas en una sola

df_all_streaming = pd.concat([df_amazon, df_disney, df_hulu, df_netflix], ignore_index = True)
```

```python
# Se crea un archivo .csv con la nueva tabla

df_all_streaming.to_csv('./Datasets/all_streaming.csv')
```
